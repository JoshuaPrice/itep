```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frames = []
for chrNum in range(1,20):
    pairsChrNum = pd.read_csv('/data2/josh/ep/activePairixChr' + str(chrNum) + '.csv')
    frames.append(pairsChrNum)
    
frames.append(pd.read_csv('/data2/josh/ep/activePairixChrX.csv'))
mergedPairs = pd.concat(frames)
```

```python
mergedPairs.reset_index(drop=True)
```

```python
sns.countplot(x='tad_status',data=mergedPairs)
plt.savefig('ep_count_all.png')
```

```python
mergedPairs.to_csv('/data2/josh/ep/activePairixAll.csv')
```

```python
chr9_counts = pd.read_csv('/data2/josh/ep/activePairixAll.csv')

fig, ax = plt.subplots(figsize=[4,3])
plt.title('Counts of Intra-, Inter-, and TAD-less EPIs in Sample')
sns.countplot(x='tad_status',data=chr9_counts)
plt.xlabel('Interaction Category')
plt.ylabel('Count in Sample')
```

```python
chr9_counts = pd.read_csv('/data2/josh/ep/activePairixAll.csv')
chr9_counts['tad_status'].value_counts()
```

```python
count = chr9_counts
inter_count = count[count['tad_status']=='Inter']
intra_count = count[count['tad_status']=='Intra']
```

```python
inter_count.sort_values('distance', ascending=False)
```

```python
inter_count['distance'].max()
```

```python
fig, ax = plt.subplots(figsize=[4,3])
sns.distplot(intra_count['distance'],kde=False,label='Intra-TAD Contact')
sns.distplot(inter_count['distance'],kde=False,label='Inter-TAD Contact')
plt.xlim([0,200000])
plt.legend()
plt.title('Distributions of Contact Distances')
plt.xlabel('Contact Distance (bp)')
plt.ylabel('Count in Sample')
```
